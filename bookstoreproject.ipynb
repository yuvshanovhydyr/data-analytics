```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

books = []

# ONLY grab actual book rows
rows = soup.find_all("tr", class_="book")

for row in rows:
    cols = row.find_all("td")

    books.append({
        "Title": cols[1].get_text(strip=True),
        "Author": cols[2].get_text(strip=True),
        "Pages": int(cols[4].get_text(strip=True)),
        "Topic": cols[5].get_text(strip=True),
        "Price": float(cols[6].get_text(strip=True).replace("$", ""))
    })

df = pd.DataFrame(books)

# ---- Analysis ----
most_books_author = df["Author"].value_counts().idxmax()
most_popular_topic = df["Topic"].value_counts().idxmax()
most_expensive_topic = df.groupby("Topic")["Price"].mean().idxmax()
most_pages_topic = df.groupby("Topic")["Pages"].mean().idxmax()

# ---- Output ----
print("Author with most books:", most_books_author)
print("Most popular topic:", most_popular_topic)
print("Most expensive topic:", most_expensive_topic)
print("Most pages topic:", most_pages_topic)
```
